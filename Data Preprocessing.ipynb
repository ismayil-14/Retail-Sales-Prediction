**Predicting the null values in Markdown 1 to 5 in the dataframe**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
df_copy= df.copy() 
```

```python
def ml_regression(df, null_features, label):
    df = df.drop(columns=null_features)

    # filter null values into separate dataframe and reset index order
    df_null = df[df[label].isnull()]
    df_null.reset_index(drop=True, inplace=True)
    print("1.null shape",df_null.shape)

    # also filter non-null values into separate dataframe and reset index order
    df_notnull = df[df[label].notnull()]
    df_notnull.reset_index(drop=True, inplace=True)
    print("2.not null ",df_notnull.shape)

    # make vector(x) and label(y) variables
    x = df_notnull.drop(columns=[label], axis=1)
    y = df_notnull[label]
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

    model = RandomForestRegressor().fit(x_train, y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
    r2_train = r2_score(y_train, y_pred_train)
    r2_test = r2_score(y_test, y_pred_test)
    accuracy = {
                'R2_train' : r2_train,
                'R2_test'  : r2_test}
    print(accuracy)

    model = RandomForestRegressor().fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # evaluate model performance using metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    metrics = {'R2': r2,
              'Mean Absolute Error': mae,
              'Mean Squared Error': mse,
              'Root Mean Squared Error': rmse}
    print(metrics)

    # predict null values using model
    y_pred = model.predict(df_null.drop(columns=[label], axis=1))

    # make dataframe of predicted label value and merge with non-null values
    df_null[label] = pd.DataFrame(y_pred)
    df_final = pd.concat([df_null,df_notnull], axis=0, ignore_index=True)
    return df_final
```

**By using Random Forest Algorithm predicting each markdown seperately**

```python
M1 = ml_regression(df,["MarkDown2","MarkDown3","MarkDown4","MarkDown5","Markdown","Date","CPI","Unemployment","Weekly_Sales"],"MarkDown1")
M2 = ml_regression(df,["MarkDown1","MarkDown3","MarkDown4","MarkDown5","Markdown","Date","CPI","Unemployment","Weekly_Sales"],"MarkDown2")
M3 = ml_regression(df,["MarkDown2","MarkDown1","MarkDown4","MarkDown5","Markdown","Date","CPI","Unemployment","Weekly_Sales"],"MarkDown3")
M4 = ml_regression(df,["MarkDown2","MarkDown3","MarkDown1","MarkDown5","Markdown","Date","CPI","Unemployment","Weekly_Sales"],"MarkDown4")
M5 = ml_regression(df,["MarkDown2","MarkDown3","MarkDown4","MarkDown1","Markdown","Date","CPI","Unemployment","Weekly_Sales"],"MarkDown5")
```
